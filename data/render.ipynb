```python
import matplotlib.pyplot as plt
import csv

timestamps = []
packets_per_second = []

with open("packet_data.csv", mode='r') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        timestamps.append(float(row[0]))
        packets_per_second.append(float(row[2]))

plt.figure(figsize=(10, 6)) 
plt.plot(timestamps, packets_per_second, linestyle='-', marker='o', color='b')

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Packets per Second', fontsize=12)
plt.title('Packets per Second Over Time', fontsize=14)

max_value = max(packets_per_second)
max_index = packets_per_second.index(max_value)

plt.show()
```
